## 🤔 Зачем нужно численное интегрирование?

### Проблема

В физике многие задачи описываются **дифференциальными уравнениями**:

$$\frac{dx}{dt} = f(x, t)$$

**Примеры:**
- Свободное падение: $\frac{dv}{dt} = -g$ ✅ *можем решить аналитически!*
- Маятник: $\frac{d^2\theta}{dt^2} = -\frac{g}{L}\sin(\theta)$ ❌ *сложно!*
- Движение планет: $\frac{d\vec{r}}{dt} = \vec{v}, \quad \frac{d\vec{v}}{dt} = -\frac{GM}{r^3}\vec{r}$ ❌ *очень сложно!*
- Турбулентность, погода, квантовая механика... ❌ *невозможно аналитически!*

### Решение

**Численное интегрирование** — заменяем производные **приближёнными формулами** и вычисляем шаг за шагом!

**Идея**:
- Разбиваем время на маленькие шаги $\Delta t$
- На каждом шаге **приближённо** вычисляем новое значение
- Чем меньше $\Delta t$, тем точнее результат

Самый простой метод — **метод Эйлера** (Leonhard Euler, 1768)

---
## 📐 Теория метода Эйлера

### Формула производной

Из определения производной:

$$\frac{dx}{dt} = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}$$

Если $\Delta t$ **маленькое, но не ноль**:

$$\frac{dx}{dt} \approx \frac{x(t + \Delta t) - x(t)}{\Delta t}$$

Откуда:

$$x(t + \Delta t) \approx x(t) + \frac{dx}{dt} \cdot \Delta t$$

### Метод Эйлера

Если известно, что $\frac{dx}{dt} = f(x, t)$, то:

$$\boxed{x_{\text{новое}} = x_{\text{старое}} + f(x_{\text{старое}}, t) \cdot \Delta t}$$

**Это и есть метод Эйлера!**

### Геометрический смысл

```
       x
       ↑
       |         • ← новая точка (Эйлер)
       |        /
       |       /  (касательная)
       |      /
       |     • ← старая точка
       |    /|
       |   / |
       |  /  | dx/dt · Δt
       | /   |
       |/____|_______→ t
           Δt
```

Мы продолжаем движение **по касательной** к графику функции!

### Для системы уравнений

Если есть несколько переменных:

$$\begin{align}
\frac{dx}{dt} &= f_x(x, y, t) \\
\frac{dy}{dt} &= f_y(x, y, t)
\end{align}$$

То применяем метод Эйлера к каждой:

$$\begin{align}
x_{\text{новое}} &= x_{\text{старое}} + f_x \cdot \Delta t \\
y_{\text{новое}} &= y_{\text{старое}} + f_y \cdot \Delta t
\end{align}$$

### Точность метода

**Локальная погрешность** (на одном шаге): $O(\Delta t^2)$

**Глобальная погрешность** (накопленная за время $T$): $O(\Delta t)$

**Вывод**: Чтобы уменьшить ошибку в 10 раз, нужно уменьшить $\Delta t$ в 10 раз!

## 🎯 Постановка задачи: Полёт камня

### Физическая постановка

Камень брошен из точки $(0, 0)$ с начальной скоростью $v_0 = 20$ м/с под углом $\theta = 45°$ к горизонту.

**Дано**:
- Начальная позиция: $(x_0, y_0) = (0, 0)$
- Начальная скорость: $v_0 = 20$ м/с
- Угол броска: $\theta = 45°$
- Ускорение свободного падения: $g = 9.81$ м/с²
- Сопротивление воздуха не учитываем

**Найти**:
1. Где камень упадёт на землю? (дальность полёта)
2. Траекторию движения $y(x)$

### Начальные условия

Разложим начальную скорость на компоненты:

$$\begin{align}
v_{x0} &= v_0 \cos(\theta) = 20 \cos(45°) = 20 \cdot \frac{\sqrt{2}}{2} \approx 14.14 \text{ м/с} \\
v_{y0} &= v_0 \sin(\theta) = 20 \sin(45°) = 20 \cdot \frac{\sqrt{2}}{2} \approx 14.14 \text{ м/с}
\end{align}$$

### Уравнения движения

**Горизонтальная ось** (нет ускорения):
$$\begin{align}
\frac{dx}{dt} &= v_x \\
\frac{dv_x}{dt} &= 0
\end{align}$$

**Вертикальная ось** (ускорение $-g$):
$$\begin{align}
\frac{dy}{dt} &= v_y \\
\frac{dv_y}{dt} &= -g
\end{align}$$

Это система из **4 дифференциальных уравнений первого порядка**!

## 💻 Численное решение методом Эйлера

Теперь решим ту же задачу **численно**, используя метод Эйлера.

````python
def simulate_projectile_euler(v0, theta_rad, g, dt, max_time=None):
    """
    Симуляция полёта снаряда методом Эйлера
    
    Обновление угла и угловой скорости
    """
    # Начальные условия
    x = 0.0
    y = 0.0
    vx = v0 * np.cos(theta_rad)
    vy = v0 * np.sin(theta_rad)
    t = 0.0
    
    # Списки для сохранения истории
    history = {
        't': [t],
        'x': [x],
        'y': [y],
        'vx': [vx],
        'vy': [vy]
    }
    
    # Симуляция
    while True:
        # Метод Эйлера:
        # x_new = x_old + vx * dt
        # y_new = y_old + vy * dt
        # vx_new = vx_old + 0 * dt  (нет ускорения по x)
        # vy_new = vy_old + (-g) * dt
        
        x += ...
        y += ...
        vx += ...  # можно опустить, но оставим для наглядности
        vy += ...
        t += ...
        
        # Сохраняем
        history['t'].append(t)
        history['x'].append(x)
        history['y'].append(y)
        history['vx'].append(vx)
        history['vy'].append(vy)
        
        # Условие остановки: упал на землю (y < 0) или достигнуто максимальное время
        if y < 0:
            break
        if max_time is not None and t >= max_time:
            break
    
    return pd.DataFrame(history)

# Запускаем симуляцию с шагом dt = 0.01 с
dt = 0.01  # шаг по времени
df_euler = simulate_projectile_euler(v0, theta_rad, g, dt)

print("="*60)
print("💻 ЧИСЛЕННОЕ РЕШЕНИЕ (метод Эйлера)")
print("="*60)
print(f"\n⚙️  Параметры симуляции:")
print(f"   Шаг по времени: dt = {dt} с")
print(f"   Количество шагов: {len(df_euler)}")

# Результаты
t_flight_euler = df_euler['t'].iloc[-1]
x_max_euler = df_euler['x'].iloc[-1]
y_max_euler = df_euler['y'].max()
t_max_height_euler = df_euler.loc[df_euler['y'].idxmax(), 't']

print(f"\n📊 Результаты симуляции:")
print(f"   Время полёта: {t_flight_euler:.4f} с")
print(f"   Дальность: {x_max_euler:.4f} м")
print(f"   Максимальная высота: {y_max_euler:.4f} м")
print(f"   (в момент времени t = {t_max_height_euler:.4f} с)")

print("\n" + "="*60)

# Первые несколько строк
print("\n📝 Первые 5 шагов симуляции:")
print(df_euler.head())

print("\n📝 Последние 5 шагов симуляции:")
print(df_euler.tail())
````

## ⚙️ Анализ точности метода: влияние шага dt

Посмотрим, как точность метода Эйлера зависит от шага по времени $\Delta t$.

````python
# Проверим разные значения dt
dt_values = [0.001, 0.01, 0.05, 0.1, 0.2, 0.5]

results = []

for dt_test in dt_values:
    df_test = simulate_projectile_euler(v0, theta_rad, g, dt_test)
    
    # Ошибки
    t_err = abs(df_test['t'].iloc[-1] - t_flight)
    x_err = abs(df_test['x'].iloc[-1] - x_max)
    y_max_test = df_test['y'].max()
    y_err = abs(y_max_test - y_max)
    
    results.append({
        'dt': dt_test,
        'N_шагов': len(df_test),
        'Ошибка t (с)': t_err,
        'Ошибка x (м)': x_err,
        'Ошибка y_max (м)': y_err,
        'Макс. ошибка (%)': max(t_err/t_flight, x_err/x_max, y_err/y_max) * 100
    })

df_accuracy = pd.DataFrame(results)

print("="*80)
print("⚙️  АНАЛИЗ ТОЧНОСТИ: зависимость от шага dt")
print("="*80)
print("\n")
print(df_accuracy.to_string(index=False))
print("\n" + "="*80)
print("\n💡 Выводы:")
print("   • Чем меньше dt, тем точнее результат")
print("   • Но меньше dt → больше шагов → дольше вычисления")
print("   • Нужен компромисс между точностью и скоростью!")
print(f"\n   Для dt = {dt} с ошибка составляет ~ {df_accuracy[df_accuracy['dt']==dt]['Макс. ошибка (%)'].values[0]:.3f}%")
````

````python
# График зависимости ошибки от dt
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_accuracy['dt'],
    y=df_accuracy['Макс. ошибка (%)'],
    mode='lines+markers',
    name='Максимальная ошибка',
    marker=dict(size=10, color='red'),
    line=dict(width=2, color='red'),
    hovertemplate='dt: %{x} с<br>Ошибка: %{y:.4f}%<extra></extra>'
))

fig.update_layout(
    title='Точность метода Эйлера в зависимости от шага dt',
    xaxis_title='Шаг времени dt (с)',
    yaxis_title='Максимальная относительная ошибка (%)',
    width=900,
    height=500,
    template='plotly_white',
    xaxis_type='log',  # логарифмическая шкала для лучшей визуализации
    yaxis_type='log'
)

fig.show()

print("📉 График показывает, что ошибка растёт линейно с ростом dt")
print("   (на log-log шкале это прямая линия с наклоном 1)")
print("\n   Это подтверждает теорию: глобальная ошибка метода Эйлера ~ O(dt)")
````
